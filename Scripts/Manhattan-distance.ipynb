# Manhattan distance calculation

The following code is for calculating "Manhattan Distance". The idea is to show the way to do it in a Jupyter Notebook and Python instead of using SQL in Hive. The initial file called `yellowtot_1.csv` comes from the preprocessed work done previously in Hive.

```python
#import libraries
import pandas as pd
import numpy as np
import string
```

```python
#read csv file preprocesed from Hive
df1 = pd.read_csv("D:/yellow-taxy-nyc/Data/yellowtot_1.csv")
df1.head()
```

```python
#rename the columns
df1.columns = [x.replace('yellowtot_1.','') for x in df1.columns]
```

```python
df1.head()
```

```python
#create manhattan distance function
def manhattan_distance(list1,list2):
    arr1 = np.asarray(list1)
    arr2 = np.asarray(list2)
    return np.sum(abs(arr1-arr2))
```

```python
#apply function to data frame
df1["pickup_lon_lat"] = df1.apply(lambda x: [x["pickup_longitude"],x["pickup_latitude"]], axis = 1)
df1["dropoff_lon_lat"] = df1.apply(lambda x: [x["dropoff_longitude"],x["dropoff_latitude"]], axis = 1)
df1["manhattan_distance"] = df1.apply(lambda x: manhattan_distance(x["pickup_lon_lat"],x["dropoff_lon_lat"]), axis = 1)
```

```python
#change columns name according to documentation
df1.columns = df1.columns.str.replace('vendorid','VendorID')
df1.columns = df1.columns.str.replace('ratecodeid','RateCodeID')
```

```python
#Save results into the folder
df1.to_csv(
    path_or_buf="D:/yellow-taxy-nyc/Data/final_dataset.csv"
    ,index= False
    ,columns=["VendorID","tpep_pickup_datetime","tpep_dropoff_datetime","passenger_count","trip_distance","pickup_longitude",
             "pickup_latitude","RateCodeID","store_and_fwd_flag","dropoff_longitude","dropoff_latitude","manhattan_distance",
             "payment_type","fare_amount","extra","mta_tax","tip_amount","tolls_amount","improvement_surcharge","total_amount"]
)
```
